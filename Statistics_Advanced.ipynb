3.  Explain the difference between discrete and continuous distributions.

Ans. A discrete distribution gives the probabilities of distinct, separate values of a discrete random variable.It assigns probability to each individual value while a continuous distribution describes the probabilities of a continuous random variable over intervals of real numbers.It assigns probability to ranges of values, not to individual points.

4. What is a binomial distribution, and how is it used in probability ?

Ans. A binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials, where each trial has only two possible outcomes (success or failure) and the probability of success remains constant.It is used in probability to calculate the likelihood of obtaining a specific number of successes in a given number of trials.

6. What is the Central Limit Theorem (CLT), and why is it critical in statistics ?

Ans. The Central Limit Theorem states that, for a sufficiently large sample size, the sampling distribution of the sample mean will be approximately normally distributed, regardless of the shape of the population distribution, provided the samples are independent and identically distributed with a finite mean and variance.The Central Limit Theorem is critical in statistics because it allows normal distribution methods to be used for inference about population means, even when the original population is not normally distributed, thereby forming the foundation for confidence intervals and hypothesis testing.

7. What is the significance of confidence intervals in statistical analysis ?

Ans. A confidence interval is a range of values, derived from sample data, that is likely to contain the true population parameter (such as the mean or proportion) with a specified level of confidence (e.g., 95%).Confidence intervals are significant in statistical analysis because they provide an estimate of the uncertainty around a sample statistic, allowing researchers to make informed inferences about the population while quantifying the reliability of their estimates.

8. What is the concept of expected value in a probability distribution ?

Ans. The expected value of a probability distribution is the long-run average or mean value of a random variable, calculated by summing (or integrating) all possible values of the variable weighted by their probabilities.The expected value is significant because it provides a measure of the central tendency of a random variable, guiding predictions and decision-making under uncertainty.

```python
''' 9. Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution. '''

import numpy as np
import matplotlib.pyplot as plt

mean = 50
std_dev = 5
n_samples = 1000

data = np.random.normal(loc=mean, scale=std_dev, size=n_samples)

computed_mean = np.mean(data)
computed_std = np.std(data)

print(f"Computed Mean: {computed_mean:.2f}")
print(f"Computed Standard Deviation: {computed_std:.2f}")

plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
''' 10.  You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

 i. Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.

ii.Write the Python code to compute the mean sales and its confidence interval.

Ans. (i.) The Central Limit Theorem (CLT) states that the sampling distribution of the sample
mean approaches a normal distribution as the sample size increases, regardless of the shape
of the original data. In the context of daily sales, the CLT allows us to treat the mean of
the collected sales data as approximately normally distributed. Using this principle, we can
estimate the true average daily sales of the company by calculating a 95% confidence interval.
This is done by computing the sample mean and sample standard deviation, then determining the
standard error of the mean by dividing the standard deviation by the square root of the sample
size. Multiplying the standard error by the z-score corresponding to a 95% confidence level gives
the margin of error, which is then added to and subtracted from the sample mean to construct the
confidence interval. This interval represents the range in which we can be 95% confident that the
actual average daily sales of the company lie. '''

import numpy as np
from scipy import stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)
n = len(daily_sales)

sem = std_sales / np.sqrt(n)

confidence_level = 0.95
z_score = stats.norm.ppf((1 + confidence_level) / 2)
margin_of_error = z_score * sem

confidence_interval = (mean_sales - margin_of_error, mean_sales + margin_of_error)

print(f"Mean Daily Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")



```
